```python
import pandas as pd
import os
```

```python
plot_info = []

data_dir = "../utils"

for root, sub_dir, files in os.walk(data_dir):
    for file in files:
        if file.endswith(".csv"):
            try:
                src_file = os.path.join(data_dir, file)
                dim = int(file[3])
                obj = int(file[9])
                tree = file[16:].strip(".csv")
                df = pd.read_csv(src_file, index_col=0).fillna(0)
                df["dimension"] = dim
                df["n_objectives"] = obj
                df["tree"] = tree
                plot_info.append(df)
            except Exception as e:
                pass
```

```python
new_df = pd.concat(plot_info)
```

```python
new_df
```

```python
solver = "MOEAD"
dim_n_obj_combs = [
    (2, 3),
    (2, 4),
    (2, 5),
    (3, 3),
    (3, 4),
    (3, 5),
    (4, 3),
    (4, 4),
    (4, 5),
    (5, 3),
    (5, 4),
    (5, 5),
]
tree = "breadth"
active_comb = dim_n_obj_combs[-1]

dimension = active_comb[0]
n_objectives = active_comb[1]
filtered_df = new_df[
    (new_df["solver"] == solver)
    & (new_df["dimension"] == dimension)
    & (new_df["tree"] == tree)
    & (new_df["n_objectives"] == n_objectives)
]
```

```python
solvers = ["MOEAD", "NSGAII", "GDE3", "IBEA"]
dimensions = [2, 3, 4, 5]
n_objectives = [2, 3, 4, 5]
tree = "breadth"
df_info = []
for dimension in dimensions:
    for n_objective in n_objectives:
        for solver in solvers:
            filtered_df = new_df[
                (new_df["solver"] == solver)
                & (new_df["dimension"] == dimension)
                & (new_df["tree"] == tree)
                & (new_df["n_objectives"] == n_objective)
            ]
            if not filtered_df.empty:
                df_info.append(
                    {
                        "solver": solver,
                        "dimX": dimension,
                        "dimZ": n_objective,
                        "root_mean": filtered_df["root"].mean(),
                        "root_std": filtered_df["root"].std(),
                        "node_1_mean": filtered_df["node_1"].mean(),
                        "node_1_std": filtered_df["node_1"].std(),
                        "node_2_mean": filtered_df["node_2"].mean(),
                        "node_2_std": filtered_df["node_2"].std(),
                        "node_3_mean": filtered_df["node_3"].mean(),
                        "node_3_std": filtered_df["node_3"].std(),
                        "node_4_mean": filtered_df["node_4"].mean(),
                        "node_4_std": filtered_df["node_4"].std(),
                    }
                )
```

```python
stat_df = pd.DataFrame(df_info)
```

```python
def compose_latex_rows(solvers_data: list, dimension: int, n_objective: int):
    mean_string = f"$\operatorname{{dim}}X={dimension}$ & Mean "
    std_string = f"$\operatorname{{dim}}Z={n_objective}$ & Std "
    for solver_data in solvers_data:
        mean_string += f"& {solver_data['root_mean']} & {solver_data['node_1_mean']} & {solver_data['node_2_mean']} & {solver_data['node_3_mean']} & {solver_data['node_4_mean']} "
        std_string += f"& {solver_data['root_std']} & {solver_data['node_1_std']} & {solver_data['node_2_std']} & {solver_data['node_3_std']} &{solver_data['node_4_std']} "
    return mean_string + "\\\\ \n" + std_string + "\\\\ \\hline \n"
```

```python
for dimension in dimensions:
    for n_objective in n_objectives:
        res_df = stat_df[
            (stat_df["dimX"] == dimension) & (stat_df["dimZ"] == n_objective)
        ]
        res_df = res_df.round(2)
        if not res_df.empty:
            data_list = [
                res_df[res_df["solver"] == "GDE3"].iloc[0],
                res_df[res_df["solver"] == "NSGAII"].iloc[0],
                res_df[res_df["solver"] == "MOEAD"].iloc[0],
            ]
            print(
                compose_latex_rows(
                    solvers_data=data_list, dimension=dimension, n_objective=n_objective
                )
            )
#             s1 = (res_df[res_df['solver']=='GDE3'].iloc[0])
#             s2 = (res_df[res_df['solver']=='NSGAII'].iloc[0])
#             str_write = (f"$\operatorname{{dim}}X={dimension}$ & Mean & {s1['root_mean']} & {s1['node_1_mean']} & {s1['node_2_mean']} & {s1['node_3_mean']} &{s1['node_4_mean']} "
#                          f"& {s2['root_mean']} & {s2['node_1_mean']} & {s2['node_2_mean']} & {s2['node_3_mean']} &{s2['node_4_mean']} \\\\ \n"
#             f"$\operatorname{{dim}}Z={n_objective}$ & Std & {s1['root_std']} & {s1['node_1_std']} & {s1['node_2_std']} & {s1['node_3_std']} &{s1['node_4_std']}  "
#             f"& {s2['root_std']} & {s2['node_1_std']} & {s2['node_2_std']} & {s2['node_3_std']} &{s2['node_4_std']} \\\\ \hline")
#             print(str_write)
#             print("\n")
```
