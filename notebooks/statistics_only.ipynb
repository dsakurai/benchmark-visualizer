```python
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

import ptitprince as pt
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
plot_info = []

for root, sub_dir, files in os.walk("../utils"):
    for file in files:
        if file.endswith(".csv"):
            try:
                src_file = file
                dim = int(file[3])
                obj = int(file[9])
                tree = file[16:].strip(".csv")
                df = pd.read_csv(src_file,index_col=0).fillna(0)
                df["dimension"] = dim
                df["n_objectives"] = obj
                df["tree"] = tree
                plot_info.append(df)
            except Exception as e:
                pass
            
```

```python
new_df = pd.concat(plot_info)
```

```python
new_df
```

```python
solver = "GDE3"
dim_n_obj_combs = [(2,3),(2,4),(2,5),(3,3),(3,4),(3,5),(4,3),(4,4),(4,5),(5,3),(5,4),(5,5)]
tree = "breadth"
active_comb = dim_n_obj_combs[-1]

dimension = active_comb[0]
n_objectives = active_comb[1]
filtered_df = new_df[(new_df["solver"]==solver) & (new_df["dimension"]==dimension)&(new_df["tree"]==tree)&(new_df["n_objectives"]==n_objectives)]
        
```

```python
solvers = ["MOEAD","NSGAII","GDE3","IBEA"]
dimensions = [2,3,4,5]
n_objectives = [2,3,4,5]
tree = "depth"
df_info = []
for dimension in dimensions:
    for n_objective in n_objectives:
        for solver in solvers:
            filtered_df = new_df[(new_df["solver"]==solver) & (new_df["dimension"]==dimension)&(new_df["tree"]==tree)&(new_df["n_objectives"]==n_objective)]
            if not filtered_df.empty:
                df_info.append({
                    "solver":solver,
                    "dimX":dimension,
                    "dimZ":n_objective,
                    "root_mean":filtered_df["root"].mean(),
                    "root_std":filtered_df["root"].std(),
                    "node_1_mean":filtered_df["node_1"].mean(),
                    "node_1_std":filtered_df["node_1"].std(),
                    "node_2_mean":filtered_df["node_2"].mean(),
                    "node_2_std":filtered_df["node_2"].std(),
                    "node_3_mean":filtered_df["node_3"].mean(),
                    "node_3_std":filtered_df["node_3"].std(),
                    "node_4_mean":filtered_df["node_4"].mean(),
                    "node_4_std":filtered_df["node_4"].std(),
                })
            
```

```python
stat_df = pd.DataFrame(df_info)
```
