GD

```python
from jmetal.core.quality_indicator import GenerationalDistance

# 真のPareto frontの取得
reference_front = all_fronts

# Generational Distanceの計算
gd = GenerationalDistance(reference_front)

front = generation_points_y

gd_value = gd.compute(front)

print("Generational Distance: ", gd_value)
```

GD for each node

```python
for node_id in pareto_dict.keys():
    
    # 真のPareto frontの取得
    reference_front = pareto_dict[node_id]["pareto_front"]

    # Generational Distanceの計算
    gd = GenerationalDistance(reference_front)

    front = generation_points_node_y[generation_points_node_y[:,2] == node_id,:][:,0:2]

    gd_value = gd.compute(front)

    print(f"Generational Distance for node[{node_id}]: ", gd_value)
```

GDX

```python
from jmetal.core.quality_indicator import GenerationalDistance

# 真のPareto setの取得
reference_set = all_sets

# Generational Distanceの計算
gd = GenerationalDistance(reference_set)

set = generation_points_x

gd_value = gd.compute(set)

print("Generational Distance X: ", gd_value)
```

GDX for each node

```python
for node_id in pareto_dict.keys():
    
    # 真のPareto setの取得
    reference_set = pareto_dict[node_id]["pareto_set"]

    # Generational Distanceの計算
    gd = GenerationalDistance(reference_set)

    set = generation_points_node_x[generation_points_node_x[:,4] == node_id,:][:,0:4]

    gd_value = gd.compute(set)

    print(f"Generational Distance X for node[{node_id}]: ", gd_value)
```

IGD

```python
from jmetal.core.quality_indicator import InvertedGenerationalDistance

# 真のPareto frontの取得
reference_front = all_fronts

# Inverted Generational Distanceの計算
igd = InvertedGenerationalDistance(reference_front)

front = generation_points_y

igd_value = igd.compute(front)

print("Inverted Generational Distance: ", igd_value)
```

IGD for each node

```python
for node_id in pareto_dict.keys():
    
    # 真のPareto frontの取得
    reference_front = pareto_dict[node_id]["pareto_front"]

    # Inverted Generational Distanceの計算
    igd = InvertedGenerationalDistance(reference_front)

    front = generation_points_node_y[generation_points_node_y[:,2] == node_id,:][:,0:2]

    # frontが空でない場合のみIGDを計算
    if front.size > 0:
        igd_value = igd.compute(front)
        print(f"Inverted Generational Distance for node[{node_id}]: ", igd_value)
    else:
        print(f"Inverted Generational Distance for node[{node_id}]: NaN")
```
