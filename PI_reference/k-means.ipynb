```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
```

```python
data_csv = "/local/pareto_benchmark/benchmark-visualizer/data/pop100_50000iter/exp_csvs/NSGAII_breadth_base_1_2_StoppingByEvaluations_2023-03-22T10-51-32.374672.csv"
X_org = pd.read_csv(data_csv, index_col=0)
```

```python
X_org.head(20)
```

```python
variable_names = [f"x{x+1}" for x in range(7)]
variable_names.insert(0, "t")
print(variable_names)
X = X_org[variable_names].values
Y = X_org[["y1", "y2"]].values

# Create a KMeans object with 3 clusters
kmeans = KMeans(n_clusters=5)

# Fit the data to the KMeans model
kmeans.fit(X)

# Print the centroids of each cluster
print(kmeans.cluster_centers_)
print(kmeans.labels_)

kmeans_y = KMeans(n_clusters=5)
kmeans_y.fit(Y)

print(kmeans_y.cluster_centers_)
print(kmeans_y.labels_)
```
