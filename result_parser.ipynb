```python
import pandas as pd
import os
from tqdm import tqdm
```

```python
def match_file(solver_name):
    file_list = []
    for root, sub_dir, files in os.walk("."):
        if root == ".":
            for file in files:
                if file.endswith(".csv"):
                    if file.split("__")[1].startswith(solver_name):
                        file_list.append(file)
    return file_list
```

```python
solvers = ["IBEA", "MOEAD", "NSGAII", "GDE3"]
n_generation = 400
pop_size = 100
step_start = (n_generation - 1) * pop_size

res_list = []
for solver in solvers:
    fs = match_file(solver_name=solver)
    for i, file in enumerate(fs):
        index = file.split("_")[2]
        eval_log = pd.read_csv(file, index_col=0)
        eval_log[["eval_node_id", "step"]] = eval_log[["eval_node_id", "step"]].astype(int)
        eval_log = eval_log[step_start : step_start + pop_size]
        result = eval_log["eval_node_id"].value_counts().to_dict()
        res_list.append(
            {   "solver":solver,
                "exp_index": i + 1,
                "root": result[0] if result.get(0) else 0,
                "node_1": result[1] if result.get(1) else 0,
                "node_2": result[2] if result.get(2) else 0,
                "node_3": result[3] if result.get(3) else 0,
                "node_4": result[4] if result.get(4) else 0,
            }
        )
print(res_list)
    

```

```python
ddf = pd.DataFrame(res_list)
```

```python
ddf.to_csv("summarization_100_exps.csv",index=False)
```
